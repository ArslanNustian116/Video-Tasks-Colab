**Task 1 Create Buttons and Display it**

```python
# Importing required modules
import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to handle button click
def on_button_clicked(button):
    clear_output(wait=True)
    if button.description == '1':
        print("1 Run")
    elif button.description == '2':
        print("2 Runs")
    elif button.description == '3':
        print("3 Runs")
    elif button.description == '4':
        print("Four Runs")
    elif button.description == '5':
        print("5 Runs")
    elif button.description == '6':
        print("Six Runs")

# Creating buttons
buttons = [widgets.Button(description=str(i)) for i in range(1, 7)]

# Displaying buttons and setting the event
for button in buttons:
    display(button)
    button.on_click(on_button_clicked)


```

**Task 2 Create an Exit Button. The Scoring Stops when Exit is Pressed**


```python

import ipywidgets as widgets
from IPython.display import display, clear_output
import os

# Filename to store the outputs
filename = "output.txt"

# Function to handle button click
def on_button_clicked(button):
    with open(filename, "a") as file:
        if button.description == '1':
            file.write("1 Run\n")
        elif button.description == '2':
            file.write("2 Runs\n")
        elif button.description == '3':
            file.write("3 Runs\n")
        elif button.description == '4':
            file.write("Four Runs\n")
        elif button.description == '5':
            file.write("5 Runs\n")
        elif button.description == '6':
            file.write("Six Runs\n")

# Function to handle exit and download
def on_exit_clicked(button):
    with open(filename, "r") as file:
        content = file.read()
    clear_output(wait=True)
    print(content)
    print("Download the file:", filename)
    from google.colab import files
    files.download(filename)

# Creating buttons
buttons = [widgets.Button(description=str(i)) for i in range(1, 7)]
exit_button = widgets.Button(description="Exit")

# Displaying buttons and setting the event
for button in buttons:
    display(button)
    button.on_click(on_button_clicked)

display(exit_button)
exit_button.on_click(on_exit_clicked)

```

**Task 3 Display Batsman, Bowler and Event That happened**


```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import os

# Filename to store the outputs
filename = "cricket_output.txt"

# Create text widgets for Bowler, Batsman, and Over Number
bowler_textbox = widgets.Text(description="Bowler:")
batsman_textbox = widgets.Text(description="Batsman:")
over_textbox = widgets.Text(description="Over No.:")

# Function to handle button click
def on_button_clicked(button):
    bowler = bowler_textbox.value
    batsman = batsman_textbox.value
    over = over_textbox.value

    with open(filename, "a") as file:
        run_text = {
            '0': "No Run",
            '1': "1 Run",
            '2': "2 Runs",
            '3': "3 Runs",
            '4': "Four Runs",
            '5': "5 Runs",
            '6': "Six Runs"
        }.get(button.description, "")
        file.write(f"{over} {bowler} to {batsman} {run_text}\n")

# Function to handle exit and download
def on_exit_clicked(button):
    with open(filename, "r") as file:
        content = file.read()
    clear_output(wait=True)
    print(content)
    print("Download the file:", filename)
    from google.colab import files
    files.download(filename)

# Creating buttons
buttons = [widgets.Button(description=str(i)) for i in range(0, 7)]
exit_button = widgets.Button(description="Exit")

# Displaying text widgets and buttons
display(bowler_textbox)
display(batsman_textbox)
display(over_textbox)

for button in buttons:
    display(button)
    button.on_click(on_button_clicked)

display(exit_button)
exit_button.on_click(on_exit_clicked)



```
